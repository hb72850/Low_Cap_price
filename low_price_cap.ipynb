```python
plt.figure(figsize = (10,6))
sns.boxplot(train['Low_Cap_Price'])
plt.title('Outlier')
plt.show()
```

```python
def replace_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    Lower_range = Q1-(1.5*IQR)
    Upper_range = Q3+(1.5*IQR)
    return Lower_range,Upper_range
```

```python
Lower_price,Upper_price= replace_outlier(train['Low_Cap_Price'])
```

```python
Lower_price
```

```python
Upper_price
```

```python
train['Low_Cap_Price']
```

```python
train['Low_Cap_Price'] = np.where(train['Low_Cap_Price']>Upper_price,Upper_price,train['Low_Cap_Price'])
train['Low_Cap_Price'] = np.where(train['Low_Cap_Price']<Lower_price,Lower_price,train['Low_Cap_Price'])
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(train['Low_Cap_Price'])
plt.title('After Removing Outlier')
plt.show()
```
